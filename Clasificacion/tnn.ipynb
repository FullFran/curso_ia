```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import make_circles

import torch
import torch.nn as nn
import torch.nn.functional as F

import torch.optim as optim

mish=nn.Mish()
relu=nn.ReLU()
```

# Clasificaci√≥n con torch

Vamos a generar datos de dos circulos con ruido y vamos a utilizar una red neuronal creada en torch para separar los datos de los dos circulos.

```python
# Creamos los datos
n=500
entry_params=2

X,Y=make_circles(n_samples=n,factor=0.5, noise=0.05)

Y=Y[:,np.newaxis]

plt.scatter(X[:,0], X[:,1])
plt.title("Puntos del dataset")
plt.show()

#Queremos separar los dos circulos as√≠:
plt.scatter(X[Y[:,0] == 0,0], X[Y[:,0] == 0,1], c='b')
plt.scatter(X[Y[:,0] == 1,0], X[Y[:,0] == 1,1], c='r')
plt.title("Como queremos que separe")
plt.show()
```

```python
# creamos nuestra red neuronal
class Net(nn.Module):

    def __init__(self):
        super(Net, self).__init__()
        
        self.layer1 = nn.Linear(2,4)
        self.layer2 = nn.Linear(4,8)
        self.layer3 = nn.Linear(8,16)
        self.layer4 = nn.Linear(16,8)
        self.layer5 = nn.Linear(8,4)
        self.layer6 = nn.Linear(4,1)

    def forward(self, x):
        x = relu(self.layer1(x))
        x = relu(self.layer2(x))
        x = relu(self.layer3(x))
        x = relu(self.layer4(x))
        x = relu(self.layer5(x))
        x = self.layer6(x)
        return x

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print('Using {} device'.format(device))
model = Net().to(device)
print(model)

# create your optimizer
optimizer = optim.SGD(model.parameters(), lr=0.03)

criterion = nn.MSELoss()
```

```python
# Transformamos los datos que tenemos en numpy a tensores de torch:
# Trasformar de numpy a Tensor
X_np = torch.from_numpy(X).float().to(device)
Y_np = torch.from_numpy(Y).float().to(device)
```

```python
for i in range(20000):
  result = model(X_np)
  #soft = nn.Softmax(dim=1)(result)
  #print(soft.shape)

  loss = criterion(result, Y_np)
  if i % 100 == 0:
    print("ERROR: ", loss.item(), "EPOCH: ", i)


  model.zero_grad()

  loss.backward()
  """
  print(model.layer1.bias.grad)
  print(model.layer2.bias.grad)
  print(model.layer3.bias.grad)
  print(model.layer4.bias.grad)
  print(model.layer5.bias.grad)
  print(model.layer6.bias.grad)
  """
  optimizer.step()

  np_result = result.cpu().detach().numpy()
  if i != 0 and i % 500 == 0:
    plt.scatter(X[np_result[:,0] < 0.5,0], X[np_result[:,0] < 0.5,1], c='skyblue')
    plt.scatter(X[np_result[:,0] > 0.5,0], X[np_result[:,0] > 0.5,1], c='salmon')
    plt.show()
  if i != 0 and i % 19999 == 0:
    print(result)

  if loss.item() < 0.0015:
    break
```
