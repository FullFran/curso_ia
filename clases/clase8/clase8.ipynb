20. Cree un .pdf que guarde un título (ej.: arial 36) y debajo una imagen que abarque
    buena parte de la página. Después:

a) Añada una página en blanco

b) Añada una página en la que se añadan, al menos, 3 líneas de texto normal

c) guarde el pdf


```python
import PyPDF2 as pdf2
from fpdf import FPDF
```

```python
# Creamos un objeto pdf y le añadimos una página
pdf = FPDF(orientation='P', unit='mm', format='A4')
pdf.add_page()
pdf.set_margins(0, 0, 0)  # Aquí seteamos los márgenes del documento

# Añadimos el título con la fuente que toca
pdf.set_font("Arial", size=36)
pdf.cell(200, 20, txt="Título", ln=1, align="C")

# Añadimos una imagen
pdf.image(w=pdf.w,
          name=r'C:\Users\Francisco\Desktop\curso_ia\derivadasbackpropagation.PNG')

# Añadimos más texto.
pdf.set_font("Arial", size=12)
pdf.cell(200, 5, txt='Figura(1): Esta es una imagen con las derivadas.',
         ln=1, align='R')
pdf.cell(200, 5, txt='Una linea de texto', ln=1)
pdf.cell(200, 5, txt='Otra linea de texto.', ln=1)

# Así guardamos el pdf.
pdf.output('prueba.pdf', 'F')
```

Para trabajar con .json necesitamos el modulo json, se usan json.dumps() y json.loads() uno lo guarda y el otro lo lee.


```python
# Vamos a trabajar con json
```

json lo que hace es convertir archivos json en diccionarios de python y diccionarios de python
en archivos .json


```python
import json
```

```python
dct = {'nombre': 'Juan', 'edad': 20, 'nacimiento': 2003,
       'coches': ['ford focus', 'opel corsa']}

json_ex = json.dumps(obj=dct)  # así convertimos un diccionario en un .json

with open('json_ex.json', 'w') as f:
    f.write(json_ex)
```

```python
with open('json_ex.json', 'r') as f:
    content = f.read()

json_ex = json.loads(content)  # así convertimos un .json en un diccionario.

print(type(json_ex))
```
