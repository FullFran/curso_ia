# Inteligencia Artificial:


Resultado de cada capa:

$\sum_jW_{ij}^cX_j^c+b_i^c = z_i^c$

Con ese resultado se realiza el fordward pass:

$X_i^{c+1} = f_i^c(z_i^c)$

Osea se repite el proceso convirtiendo los outputs de

la capa anterior en los inputs de la siguiente.

las 'f' funciones de activacin tienen que:

no ser lineales, elegir el rango de la funcin,

la funcin debe ser continuamente diferenciable.

- Funcin identidad: no quita linealidad: f(x) = x se usa para que
  la salida de la capa de entrada es lo que le has metido en la entrada.

- ReLu: cero si x<0 y x si x>0 (max(0,x)) (no diferenciable en 0, apaga neuronas...)

- Sigmoide $f(x)= (1+e^{-z})^{-1}$ tiene rango entre 0 y 1 desprecia valores atpicos y clasifica valores.

- Binary step (0 o 1, redondeo de sigmoid)

- Tangente hiperblica $\frac{e^x-e^{-x}}{e^x+e^{-x}}$ va entre -1 y 1

- Silu: sigmoid(x)\*x, ELU, GeLu...

- Hay otras muy usadas: funcin Gaussiana de to la vida, softplus (logaritmica $ln(e^x-1)$), la softmax se usa mucho, no tiene expresin analitica, se saca como desrrollo en sumatorio.


Una vez definida la red tenemos que definir una funcin para entrenarla que depender de los hiperparmetros como:

* Learning rate: 'lo rpido o lento que aprende el casharro'

* Prdida (cost): lo que difieren nuestros resultados de los resultados reales. Lo ms normal es utilizar el error cuadrtico medio o la binary cross entropy.

## Backpropagation:

Una vez calculada la prdida, queremos adaptar los pesos para que nos de un mejor resultado.

* descenso del gradiente.

* descenso del gradiente con momento: se tiene en cuenta el cambio anterior (es como dejar caer una bola rodando por la pendiente de una colina.)
(as podemos evitar mnimos locales ms fcilmente!!!!)

* Regularizacin: evita que los pesos se hagan muy muy grandes y as ocurra overfitting, utilizando tcnicas como la regresin, funciones de activacin anteriores para hacer 'filtros'

* Dropout: desactivar pesos al azar en cada paso del entrenamiento para evitar que 'solo se activen ciertos caminos' y as evitar overfitting. (se desactiva en cada paso una serie de conexiones al azar (caminos))

* aprendizaje por lotes (Batch learning): se une en cada iteracin del entrenamiento un conjunto de datos a la vez.

El objetivo del entrenamiento es reducir al mnimo el error.

* al entrenar se suele dividir los datos en (train, valid, test), la parte de train ajusta los pesos, la parte de valid ajusta los hiperparmetros y la parte de test comprueba los resultados obtenidos.
