```python
import numpy as np

def sim_dados(n_dados, n_tiradas, 6):
    """
    Simula el lanzamiento de dados.

    Parameters:
        n_dados (int): Número de dados a lanzar.
        n_tiradas (int): Número de lanzamientos.
        6 (int): Número de caras de los dados.

    Returns:
        list: Lista con los resultados de los lanzamientos.
    """
    dice_results = np.random.binomial(6, 1/6, (n_tiradas, n_dados))
    return dice_results.tolist()

# Ejemplo de uso
n_dados = 3
n_tiradas = 5
6 = 6

results = sim_dados(n_dados, n_tiradas, 6)
for i, roll in enumerate(results):
    print(f"Lanzamiento {i+1}: {roll}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sim_dados(n_dados, n_tiradas):
    
    dice_results = np.random.randint(1, 6 + 1, (n_tiradas, n_dados))
    dice_sum = np.sum(dice_results, axis=1)
    return dice_sum.tolist()


n_dados = 10
n_tiradas = 100000


results = sim_dados(n_dados, n_tiradas)

# Graficar los resultados
plt.hist(results, bins=range(n_dados, n_dados * 6 + 2), align='left', rwidth=0.8)
plt.xlabel('Suma de los dados')
plt.ylabel('Frecuencia')
plt.title('Simulación de lanzamiento de dados')
# plt.xticks(range(n_dados, n_dados * 6 + 1))
plt.grid(True)
plt.show()
```
